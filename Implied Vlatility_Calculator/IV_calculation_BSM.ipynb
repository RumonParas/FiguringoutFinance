```python
import numpy as np
import pandas as pd
import yfinance as yf
from scipy.stats import norm
```

```python
def d1(S,K,r,Stdev,T):
  return (np.log(S/K) + (r + Stdev**2/2)*T)/(Stdev*np.sqrt(T))
```

```python
def d2(S,K,r,Stdev,T):
  return (np.log(S/K) + (r - Stdev**2/2)*T)/(Stdev*np.sqrt(T))
```

```python
def BSM(S,K,r,Stdev,T):
  return (S*norm.cdf(d1(S,K,r,Stdev,T))) - (K*np.exp(-r*T)*norm.cdf(d2(S,K,r,Stdev,T)))
```

```python
data = yf.download('^NSEI', start = '2020-01-01')['Close']
data
```

```python
S = data.iloc[-1]
S
```

```python
r = 0.02
K = 23000
T = 5/365
#Std_dev = 0.29
```

```python
market_price = 386.6 # call option price at strike

#We will start sigma at 10% and keep increasing it till we solve the equation
sigma_test = 0.1
model_price = 0

while model_price < market_price:
  sigma_test = sigma_test + 0.001
  model_price = BSM(S,K,r,sigma_test,T)[0]

print(f"The implied volatility is {sigma_test*100}%")
```
